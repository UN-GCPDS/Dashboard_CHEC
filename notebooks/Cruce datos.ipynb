```python
# @title Base de datos Eventos
# Cargar bases de datos de eventos
eventos = pd.read_csv('/content/drive/Shareddrives/CHEC/data_chec/datos/Data_Total/EVENTOS*.csv')

eventos['inicio'] = pd.to_datetime(eventos['inicio']).dt.normalize()
eventos['fin'] = pd.to_datetime(eventos['fin']).dt.normalize()
eventos['inicio_m'] = pd.to_datetime(eventos['inicio']).dt.to_period('M')
eventos['fin_m'] = pd.to_datetime(eventos['fin']).dt.to_period('M')
eventos[['duracion_h','CNT_TRAFOS_AFEC','cnt_usus','SAIDI','SAIFI']] = eventos[['duracion_h','CNT_TRAFOS_AFEC','cnt_usus','SAIDI','SAIFI']].astype('float32')

# Crear un DataFrame con columnas, tipos de datos y valores nulos
info = pd.concat([eventos.dtypes.astype(str), eventos.isnull().sum()], axis=1, keys=['Tipo de Dato', 'Valores Nulos'])

# Formatear la salida
for col, dtype, nulls in zip(info.index, info['Tipo de Dato'], info['Valores Nulos']):
    print(f"{col:<12} {dtype:<10} {nulls}")

# Lista de columnas a excluir
columnas_a_excluir = ['evento', 'equipo_ope','cto_equi_ope','causa','inicio_m','fin_m']

# Obtener todas las columnas excepto las de la lista de exclusión
eventos_columnas = [col for col in eventos.columns if col not in columnas_a_excluir]

eventos
```

```python
# @title Base de datos Trafos
# Cargar bases de datos de Trafos
trafos = pd.read_csv('/content/drive/Shareddrives/CHEC/data_chec/datos/Data_Total/TRAFOS*.csv')

trafos['FECHA'] = pd.to_datetime(trafos['FECHA']).dt.to_period('M')
trafos[['PHASES','IMPEDANCE*']] = trafos[['PHASES','IMPEDANCE*']].astype('float32')
trafos[['TIPO_SUB*','GRUPO015']] = trafos[['TIPO_SUB*','GRUPO015']].astype('object')

# Crear un DataFrame con columnas, tipos de datos y valores nulos
info = pd.concat([trafos.dtypes.astype(str), trafos.isnull().sum()], axis=1, keys=['Tipo de Dato', 'Valores Nulos'])

# Formatear la salida
for col, dtype, nulls in zip(info.index, info['Tipo de Dato'], info['Valores Nulos']):
    print(f"{col:<12} {dtype:<10} {nulls}")

# Lista de columnas a excluir
columnas_a_excluir = ['CODE','FPARENT','TRFTYPE','ELNODE','CASO']

# Obtener todas las columnas excepto las de la lista de exclusión
trafos_columnas = [col for col in trafos.columns if col not in columnas_a_excluir]

trafos
```

```python
# @title Base de datos Switches
# Cargar bases de datos de switches
switches = pd.read_csv('/content/drive/Shareddrives/CHEC/data_chec/datos/Data_Total/SWITCHES*.csv')

switches['FECHA'] = pd.to_datetime(switches['FECHA']).dt.to_period('M')
switches['PHASES'] = switches['PHASES'].astype('float32')
switches['KV'] = switches['KV'].astype('object')
switches['STATE'] = switches['STATE'].astype('object')

# Crear un DataFrame con columnas, tipos de datos y valores nulos
info = pd.concat([switches.dtypes.astype(str), switches.isnull().sum()], axis=1, keys=['Tipo de Dato', 'Valores Nulos'])

# Formatear la salida
for col, dtype, nulls in zip(info.index, info['Tipo de Dato'], info['Valores Nulos']):
    print(f"{col:<12} {dtype:<10} {nulls}")

# Lista de columnas a excluir
columnas_a_excluir = ['CODE','ASSEMBLY','FPARENT','LINESECTIO']

# Obtener todas las columnas excepto las de la lista de exclusión
switches_columnas = [col for col in switches.columns if col not in columnas_a_excluir]

switches
```

```python
# @title Cruce de datos eventos con trafos
eventos_trafos = eventos[eventos['tipo_equi_ope']=='transformador']
print(eventos_trafos.shape)
eventos_trafos = pd.merge(eventos_trafos, trafos, left_on=['equipo_ope', 'inicio_m'], right_on=['CODE', 'FECHA'], how='left')
eventos_trafos.to_csv('/content/drive/Shareddrives/CHEC/data_chec/datos/Data_Total/EVENTOS_TRAFOS.csv', index=False)
eventos_trafos
```

```python
# @title
# Crear un DataFrame con el tipo de dato, valores nulos y valores únicos
info = pd.concat([
    eventos_trafos.dtypes.astype(str),                     # Tipo de dato
    eventos_trafos.isnull().sum(),                         # Valores nulos
    eventos_trafos.nunique()                               # Número de valores únicos
], axis=1, keys=['Tipo de Dato', 'Valores Nulos', 'Valores Únicos'])

# Lista para almacenar las columnas con menos o iguales a 10 valores únicos
columnas_pocas_unicas = []

# Formatear la salida e identificar columnas con <= 10 valores únicos
for col, dtype, nulls, uniques in zip(info.index, info['Tipo de Dato'], info['Valores Nulos'], info['Valores Únicos']):
    print(f"{col:<12} {dtype:<10} {nulls:<15} {uniques}")

    # Añadir a la lista si tiene <= 10 valores únicos
    if uniques <= 10:
        columnas_pocas_unicas.append(col)

# Mostrar la lista resultante
print("Columnas con menos o iguales a 10 valores únicos:", columnas_pocas_unicas)
```

```python
# @title Cruce de datos eventos con switches
eventos_switches = eventos[eventos['tipo_equi_ope']=='interruptor']
print(eventos_switches.shape)
eventos_switches = pd.merge(eventos_switches, switches, left_on=['equipo_ope', 'inicio_m'], right_on=['CODE', 'FECHA'], how='left')
eventos_switches.drop_duplicates(inplace=True)
eventos_switches.to_csv('/content/drive/Shareddrives/CHEC/data_chec/datos/Data_Total/EVENTOS_SWITCHES.csv', index=False)
eventos_switches
```

```python
# @title
# Crear un DataFrame con el tipo de dato, valores nulos y valores únicos
info = pd.concat([
    eventos_switches.dtypes.astype(str),                     # Tipo de dato
    eventos_switches.isnull().sum(),                         # Valores nulos
    eventos_switches.nunique()                               # Número de valores únicos
], axis=1, keys=['Tipo de Dato', 'Valores Nulos', 'Valores Únicos'])

# Lista para almacenar las columnas con menos o iguales a 10 valores únicos
columnas_pocas_unicas = []

# Formatear la salida e identificar columnas con <= 10 valores únicos
for col, dtype, nulls, uniques in zip(info.index, info['Tipo de Dato'], info['Valores Nulos'], info['Valores Únicos']):
    print(f"{col:<12} {dtype:<10} {nulls:<15} {uniques}")

    # Añadir a la lista si tiene <= 10 valores únicos
    if uniques <= 10:
        columnas_pocas_unicas.append(col)

# Mostrar la lista resultante
print("Columnas con menos o iguales a 10 valores únicos:", columnas_pocas_unicas)
```
